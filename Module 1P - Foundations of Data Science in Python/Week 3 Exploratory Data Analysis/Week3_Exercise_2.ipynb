#### Import Libraries & Dataset

```python
import numpy
import pandas
%matplotlib inline
import matplotlib.pyplot as plt

# import the marathon dataset
data = pandas.read_csv('marathon-data.csv')
data.head()

# transform format"hh:mm:ss" to seconds
def time_convert(x):
    h,m,s = map(int,x.split(':'))
    return (h*60+m)*60+s

data['split'] = data['split'].apply(time_convert)
data['final'] = data['final'].apply(time_convert)
```
